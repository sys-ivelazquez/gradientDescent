# Modules

```python
import numpy as np #to perform calculations
import matplotlib.pyplot as plt #to plot the points
from  mpl_toolkits.mplot3d import Axes3D #to plot 3D
```

# Functions

```python
# Function to be optimized
def mathfunc(x, y):
    return x**4 - y**2

# Derivate of mathfunc function with respect of x and y
def dxmathfunc(x, y):
    return 4*x**3

def dymathfunc(x, y):
    return -2*y

# Gradient Descent algorithm
def gdalgorithm(learningRate, iterations, x_start, y_start):
    x, y = x_start, y_start

    # Iteration progress lists
    iterationList = []
    iterationList2 = []
    x_values, y_values = [], []

    # Iterations
    for _ in range(iterations):
        # Calculate the gradient
        x_gradient = dxmathfunc(x, y)
        y_gradient = dymathfunc(x, y)

        # Update of the x and y values opposite of the gradient
        x = x - learningRate * x_gradient
        y = y - learningRate * y_gradient

        # Update of the lists appending the new values
        iterationList.append((x, y, mathfunc(x, y)))
        iterationList2.append(mathfunc(x, y))
        x_values.append(x)
        y_values.append(y)
        
    return x, y, mathfunc(x, y), iterationList, iterationList2, x_values, y_values
```

# Parameters and call

```python
# Defining the meshgrid for plotting the function
x_range = np.arange(-10, 10, 0.1)
y_range = np.arange(-10, 10, 0.1)
x, y = np.meshgrid(x_range, y_range)
z = mathfunc(x, y)

# Parameters
x_start = 1
y_start = 1
learningRate = 0.1
iterations = 1000

# Calling the gdalgorithm function
x, y, f, iterationList, iterationList2, x_values, y_values = gdalgorithm(learningRate, iterations, x_start, y_start)
```

# Plotting 2D results

```python
# Function progress within iterations
plt.figure(figsize=(12, 6))

# First subplot with the evolution of the function
plt.subplot(1, 2, 1)
plt.plot(iterationList2, label = "Function value", color = "b")
plt.title("Function progress within iterations")
plt.xlabel("Iterations")
plt.ylabel("Function value")
plt.grid(True)

# Second subplot with the trajectory of x and y within iterations
plt.subplot(1, 2, 2)
plt.plot(x_values, y_values, label = "x and y trajectory", color = "r", marker = "o")
plt.scatter(x, y, color = "g", label = "Minimum found")
plt.title("Optimization trajectory of x and y")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Plot the results
plt.tight_layout()
plt.show()
```

# Plotting 3D results

```python
# Plot the results
figure = plt.figure()
ax = figure.add_subplot(111, projection = "3d")
ax.plot_surface(x, y, z, cmap = "coolwarm")
ax.scatter(*zip(*iterationList), c = "r", marker = "o")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("mathfunc(x, y)")
plt.show()
```
